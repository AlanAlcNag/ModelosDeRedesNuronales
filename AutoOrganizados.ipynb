```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import susi
from susi.SOMPlots import plot_nbh_dist_weight_matrix, plot_umatrix
```

# emoji

```python
emoji_data = np.loadtxt('emojis.txt')
emoji_x = emoji_data[:,1:]
emoji_y = emoji_data[:,0].astype(int)
```

```python
np.unique(emoji_y)
```

```python
som = susi.SOMClustering(n_rows=16,n_columns=16)
som.fit(emoji_x)
```

```python
clusters = som.get_clusters(emoji_x)
```

```python
plt.scatter(x=[c[1] for c in clusters], y=[c[0] for c in clusters], c=emoji_y, cmap='Set2', alpha=0.2)
plt.gca().invert_yaxis()
plt.show()
```

```python
n = 5
img_ind = som.get_datapoints_from_node((14,14))
plt.figure(figsize=(10, 4))
for i in range(n):
    # Original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(emoji_x[img_ind[i]].reshape(32, 32))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

```python
n = 5
img_ind = som.get_datapoints_from_node((2,2))
plt.figure(figsize=(10, 4))
for i in range(n):
    # Original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(emoji_x[img_ind[i]].reshape(32, 32))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

```python
u_matrix = som.get_u_matrix()
plot_umatrix(u_matrix, 16, 16)
plt.show()
```

# MNIST

```python
mnist_train = pd.read_csv('fashion-mnist_train.csv')
mnist_test = pd.read_csv('fashion-mnist_test.csv')
mnist_data = pd.concat([mnist_train, mnist_test]).to_numpy()
mnist_y = mnist_data[:,0].astype(int)
mnist_x = mnist_data[:,1:]
```

```python
mnist_som = susi.SOMClustering(n_rows=37,n_columns=37)
mnist_som.fit(mnist_x)
clusters = mnist_som.get_clusters(mnist_x)
```

```python
plt.scatter(x=[c[1] for c in clusters], y=[c[0] for c in clusters], c=mnist_y, alpha=0.2)
plt.gca().invert_yaxis()
plt.show()
```

```python
n = 5
img_ind = mnist_som.get_datapoints_from_node((30,5))
plt.figure(figsize=(10, 4))
for i in range(n):
    # Original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(mnist_x[img_ind[i]].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```
